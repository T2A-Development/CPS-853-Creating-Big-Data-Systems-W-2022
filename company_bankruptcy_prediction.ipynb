```
#Checking Null values
# This function use to print feature with null values and null count 
def null_value_count(df):
  null_columns_counts = []
  numRows = df.count()
  for k in df.columns:
    nullRows = df.where(col(k).isNull()).count()
    if(nullRows > 0):
      temp = k,nullRows
      null_columns_counts.append(temp)
  return(null_columns_counts)
```

Machine Learning

```
from pyspark.ml.feature import VectorAssembler
from pyspark.ml import Pipeline

# Convert feature columns to a numeric vector
assembler_features = VectorAssembler(inputCols=feature_cols, 
                                     outputCol='features')

# Construct and execute a pipeline, cache the results.
pipeline = Pipeline(stages=[assembler_features])
allData = pipeline.fit(df2).transform(df2).cache()
```

```
(trainingData, testData) = allData.randomSplit([0.8, 0.2], seed=420)
```

Here is the list of few Classification Algorithms from Spark ML

```
from pyspark.ml.evaluation import BinaryClassificationEvaluator
## Calculate and return the AUC metric
def evaluate(testData, predictions):
    evaluator = BinaryClassificationEvaluator(labelCol="class",
                    rawPredictionCol="rawPrediction")
    print ("AUC: %f" % evaluator.evaluate(predictions))
```

LogisticRegression

```
from pyspark.ml.classification import LogisticRegression
lr = LogisticRegression(maxIter=10, regParam=0.3,
                elasticNetParam=0.8 ,labelCol="class")
lrModel = lr.fit(trainingData)
evaluate(testData, lrModel.transform(testData))
```

DecisionTreeClassifier

```
from pyspark.ml.classification import DecisionTreeClassifier
dt = DecisionTreeClassifier(labelCol="class")
dtModel = dt.fit(trainingData)
evaluate(testData, dtModel.transform(testData))
```

RandomForestClassifier

```
#RandomForestClassifier
from pyspark.ml.classification import RandomForestClassifier as RF
from pyspark.ml.evaluation import BinaryClassificationEvaluator
rf = RF(labelCol='class', featuresCol='features',
numTrees=100, maxDepth=10)
rfModel = rf.fit(trainingData)
evaluate(testData, rfModel.transform(testData))
```

Gradient-boosted tree classifier

```
#Gradient-boosted tree classifier
from pyspark.ml.classification import GBTClassifier
gbt = GBTClassifier(labelCol="class", featuresCol="features",maxIter=10)
gbt_model = gbt.fit(trainingData)
evaluate(testData, gbt_model.transform(testData))
```

NaiveBayes

```
#NaiveBayes
from pyspark.ml.classification import NaiveBayes
nb = NaiveBayes(labelCol="class", featuresCol="features")
nb_model = nb.fit(trainingData)
evaluate(testData, nb_model.transform(testData))
```

Support Vector Machine

```
#Support Vector Machine
from pyspark.ml.classification import LinearSVC
svm = LinearSVC(labelCol="class", featuresCol="features")
svm_model = svm.fit(trainingData)
evaluate(testData, svm_model.transform(testData))
```
